# Que porcentage de ganadores tiene EEUU por decada?

Conociendo que EEUU es el pais que más ha ganado, conocemos el porcentaje de victoria que han tenido por decada.

```python
nobel["usa_born_winner"] = nobel["birth_country"] == "United States of America"
nobel.head()
```

```python
nobel["decade"] = (np.floor(nobel["year"] / 10 )*10).astype(int)
nobel.head()
```

```python
nobel["decade"] = (np.floor(nobel["year"] / 10 )*10).astype(int)
nobel.head()
```

```python
prop_usa_winners = nobel.groupby("decade",as_index=False)["usa_born_winner"].mean()
prop_usa_winners
```

```python
plt.rcParams["figure.figsize"] = [11,7]

ax=sns.lineplot(
    x="decade",
    y="usa_born_winner",
    data = prop_usa_winners
)

from matplotlib.ticker import Percentformatter

ax.yaxis.set_major_formatter(Percentformatter())
```

# Cuál es el porcentaje de victoria entre hombres y mujeres en las distintas categorias?

```python
nobel["female_winner"] = nobel["sex"] == "Female"
prop_female_winners = nobel.groupby(["decade","category"],as_index=False)["female_winner"].mean()

prop_female_winners
```

```python
plt.rcParams["figure.figsize"] = [11,7]

ax=sns.lineplot(
    x="decade",
    y="female_winner",
    data = prop_female_winners,
    hue = "category"
)

from matplotlib.ticker import Percentformatter

ax.yaxis.set_major_formatter(Percentformatter())
```

- Las categorias en donde menos mujeres han ganado son las siguientes:
 - Fisica
 - Química
 - Economia
- Las categorias en donde las mujeres han ganado más son las siguientes:
 - Literatura
 - Medicina
 - Paz


# Edad hay más ganadores

```python
nobel["birth_date"] = pd.to_datetime(nobel["birth_date"])

nobel["age"] = nobel["year"] - nobel["birth_date"].dt.year

nobel.head()
```

```python
sns.lmplot(
    x="year",
    y="age",
    data = nobel,
   # lowess=True,
    aspect=2,
    line_kws={"color":"black"}
)
```

## Conlusion
entre más tiempo pasa mas tiempo requieres para realizar nuevos conocimientos. Lo cual se ve reflejado en la edad

# Rango de edad de los ganadores en las diferentes categorias

```python
nobel["birth_date"] = pd.to_datetime(nobel["birth_date"])

nobel["age"] = nobel["year"] - nobel["birth_date"].dt.year
sns.lmplot(
    x="year",
    y="age",
    data = nobel,
   # lowess=True,
    aspect=2,
    line_kws={"color":"black"},
    row="category"
)
```

Las categorias en donde la edad necesaria para ganar el premio nobel ha ido en aumento son las siguientes:
- Fisica
- Quimica
- Medicina

y las categorias en donde el rango de edad se mantiene o ha bajado son:
- Paz
- Economia
- Literatura
